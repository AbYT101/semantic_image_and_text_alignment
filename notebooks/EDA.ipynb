```python
from PIL import Image
import os

# Function to analyze image properties
def analyze_images(folder_path):
    image_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg') or f.endswith('.png')]
    image_properties = []
    
    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        img = Image.open(img_path)
        width, height = img.size
        image_properties.append({
            'filename': img_file,
            'width': width,
            'height': height,
            'format': img.format
        })
    
    return image_properties

# Analyze images in the Assets folder
image_properties = analyze_images(assets_path)

# Print image properties for the first few images as an example
for img_prop in image_properties[:5]:
    print(img_prop)
```
